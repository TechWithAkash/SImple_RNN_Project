```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding,SimpleRNN, LSTM, Dense
```

```python
# Loading the IMDB Dataset
max_features = 10000 # Vocabulary Size
(X_train, y_train), (X_test, y_test) = imdb.load_data(num_words=max_features)

# Print the Shape fo the Data
print(f"Trainig data shape : {X_train.shape},  Training Labels shape : {y_train.shape}")
print(f"Testing data shape : {X_test.shape},  Testing Labels shape : {y_test.shape}")
```

```python
# Inspect a Sample Review and Its Label
Sample_review = X_train[0] # INput Features i.e Independent Features
Sample_label = y_train[0] # Output Features i.e Dependent Features
print(f"Sample Review : {Sample_review}")
print(f"Sample Label : {Sample_label}")
```

```python
# Mapping of Words index back to Words(For Understanding)
word_index = imdb.get_word_index()
reverse_word_index = dict([(value, key) for (key, value) in word_index.items()])
```

```python
reverse_word_index # Here in these we see that data what the Particualr index is on it store
```

```python
decoded_review = ' '.join([reverse_word_index.get(i-3, '?') for i in Sample_review])
print(f"Decoded Review : {decoded_review}")
#
```

```python
from tensorflow.keras.preprocessing import sequence
from tensorflow.keras.preprocessing.sequence import pad_sequences

max_len = 500
X_train = sequence.pad_sequences(X_train, maxlen=max_len)
X_test = sequence.pad_sequences(X_test, maxlen=max_len)
X_train
```

```python
X_train[0]
```

```python
# Train Simple RNN
model = Sequential()
model.add(Embedding(max_features, 128,input_length=max_len)) # Here it get all the Words and Convert into vectors with these 128 Dimension
model.add(SimpleRNN(128,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

```python
X_train[0]
```

```python
# Creating an Istance of EarlyStopping Callback
from tensorflow.keras.callbacks import EarlyStopping
early_stopping = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)
early_stopping
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# Traininf the Model with Early Stopping
history = model.fit(
    X_train, y_train,
    epochs=5,
    batch_size=32,
    validation_split=0.2,
    callbacks=[early_stopping]
)
```

```python
# Save the MOdels
model.save('simple_rnn_model_imdb.h5')
```

```python
# Helper Function
# Function to Decode the Review
def decoded_review(encoded_review):
  return ' '.join([reverse_word_index.get(i-3, '?') for i in encoded_review])


# Function to Preprocess the User Input
def preprocess_text(text):
  words = text.lower().split()
  encoded_review = [word_index.get(word, 2)  + 3 for word in words]
  padded_review = sequence.pad_sequences([encoded_review], maxlen=500)
  return padded_review
```

```python
# Prediction FUnction
def predict_sentiment(review):
  preprocessed_review = preprocess_text(review)
  prediction = model.predict(preprocessed_review)
  sentiment = 'Positive' if prediction[0][0] > 0.5 else 'Negative'
  return sentiment,prediction[0][0]
```

```python
# Exmaple of the Review Preidctions
example_review = "The Moview is  really bad see the children due to showing the adult content in the moview."
sentiment,score = predict_sentiment(example_review)
print(f"Review : {example_review}")
print(f"Sentiment : {sentiment}")
print(f"Score : {score}")
```
